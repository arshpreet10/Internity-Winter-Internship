```
import pandas as pd
import numpy as np
```

```
from sklearn.datasets.samples_generator import make_blobs
X, y = make_blobs(n_samples=100, centers=3, random_state=52)
```

```
import random
```

```
centroids_idx = random.sample(range(0, len(X)), 3)
centroids = np.array([X[i,:] for i in centroids_idx])
centroids
```

```
def euclidean(x1, x2):
  return(sum((x1 - x2)**2))**0.5
```

```
def assign_centroid(X, centroids):
  assigned_centroids = []
  for point in X:
    distances = [euclidean(center,point) for center in centroids]
    assigned_centroids.append(np.argmin(distances))
  return assigned_centroids
```

```
#centroid for each observation
cfeo = assign_centroid(X, centroids)
print(cfeo)
```

```
idx = [cfeo[j]==1 for j in range(len(cfeo))]
np.mean(X[idx], axis=0)
```

```
def recalculate_centroids(centroids, X):
  cfeo = assign_centroid(X, centroids)
  new_centroids = np.zeros((3,2))
  for i in range(centroids.shape[0]):
    idx = [cfeo[j]==i for j in range(len(cfeo))]
    new_centroids[i] = np.mean(X[idx], axis=0)
  return new_centroids
```
